This helper function allows a visual representation of the maze object:

```python
def show(qmaze):
    plt.grid('on')
    nrows, ncols = qmaze.maze.shape
    ax = plt.gca()
    ax.set_xticks(np.arange(0.5, nrows, 1))
    ax.set_yticks(np.arange(0.5, ncols, 1))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    canvas = np.copy(qmaze.maze)
    for row,col in qmaze.visited:
        canvas[row,col] = 0.6
    pirate_row, pirate_col, _ = qmaze.state
    canvas[pirate_row, pirate_col] = 0.3   # pirate cell
    canvas[nrows-1, ncols-1] = 0.9 # treasure cell
    img = plt.imshow(canvas, interpolation='none', cmap='gray')
    return img
```

The pirate agent can move in four directions: left, right, up, and down. 

While the agent primarily learns by experience through exploitation, often, the agent can choose to explore the environment to find previously undiscovered paths. This is called "exploration" and is defined by epsilon. This value is typically a lower value such as 0.1, which means for every ten attempts, the agent will attempt to learn by experience nine times and will randomly explore a new path one time. You are encouraged to try various values for the exploration factor and see how the algorithm performs.

```python
LEFT = 0
UP = 1
RIGHT = 2
DOWN = 3


# Exploration factor
epsilon = 0.1

# Actions dictionary
actions_dict = {
    LEFT: 'left',
    UP: 'up',
    RIGHT: 'right',
    DOWN: 'down',
}

num_actions = len(actions_dict)
```

This function simulates a full game based on the provided trained model. The other parameters include the TreasureMaze object and the starting position of the pirate.

```python
def play_game(model, qmaze, pirate_cell):
    qmaze.reset(pirate_cell)
    envstate = qmaze.observe()
    while True:
        prev_envstate = envstate
        # get next action
        q = model.predict(prev_envstate)
        action = np.argmax(q[0])

        # apply action, get rewards and new state
        envstate, reward, game_status = qmaze.act(action)
        if game_status == 'win':
            return True
        elif game_status == 'lose':
            return False
```

This function helps you to determine whether the pirate can win any game at all. If your maze is not well designed, the pirate may not win any game at all. In this case, your training would not yield any result. The provided maze in this notebook ensures that there is a path to win and you can run this method to check.

```python
def completion_check(model, qmaze):
    for cell in qmaze.free_cells:
        if not qmaze.valid_actions(cell):
            return False
        if not play_game(model, qmaze, cell):
            return False
    return True
```

The code you have been given in this block will build the neural network model. Review the code and note the number of layers, as well as the activation, optimizer, and loss functions that are used to train the model.

```python
def build_model(maze):
    model = Sequential()
    model.add(Dense(maze.size, input_shape=(maze.size,)))
    model.add(PReLU())
    model.add(Dense(maze.size))
    model.add(PReLU())
    model.add(Dense(num_actions))
    model.compile(optimizer='adam', loss='mse')
    return model
```

This is your deep Q-learning implementation. The goal of your deep Q-learning implementation is to find the best possible navigation sequence that results in reaching the treasure cell while maximizing the reward. In your implementation, you need to determine the optimal number of epochs to achieve a 100% win rate.

You will need to complete the section starting with #pseudocode. The pseudocode has been included for you.

```python
import numpy as np
import random
import datetime

# Global epsilon should exist already; if not, uncomment:
# epsilon = 1.0

def qtrain(model, maze, **opt):
    """
    Deep Q-learning training loop for the pirate agent.

    Expects:
      - TreasureMaze(maze) environment
      - GameExperience(model, max_memory=...) replay buffer with:
          remember([state, action, reward, next_state, status])
          get_data(data_size=..., gamma=...) -> (inputs, targets)
    """
    global epsilon

    # ---- Hyperparameters / options ----
    n_epoch    = opt.get('n_epoch', 15000)      # training epochs (episodes)
    max_memory = opt.get('max_memory', 1000)    # replay capacity (used by GameExperience)
    data_size  = opt.get('data_size', 50)       # batch size sampled from replay
    gamma      = opt.get('discount', 0.95)      # RL discount factor
    eps_min    = opt.get('epsilon_min', 0.10)   # minimum exploration
    eps_decay  = opt.get('epsilon_decay', 0.995)# epsilon decay per episode

    # ---- Setup ----
    start_time = datetime.datetime.now()
    qmaze = TreasureMaze(maze)
    experience = GameExperience(model, max_memory=max_memory)

    win_history = []                 # 1 for win, 0 for loss per episode
    hsize = qmaze.maze.size // 2     # window size for win-rate smoothing
    win_rate = 0.0

    for epoch in range(n_epoch):
        # Random free starting cell (target excluded by TreasureMaze)
        start_cell = random.choice(qmaze.free_cells)
        qmaze.reset(start_cell)
        envstate = qmaze.observe()
        status = 'not_over'
        loss = 0.0

        while status == 'not_over':
            # ---- ε-greedy action with valid-action masking ----
            valid_actions = qmaze.valid_actions()
            if (not valid_actions) or (np.random.rand() < epsilon):
                action = int(random.choice(valid_actions)) if valid_actions else 0
            else:
                # Exploit: predict Q(s, ·) and mask invalid actions
                q_values = model.predict(envstate, verbose=0)[0]  # shape (4,)
                masked_q = q_values.copy()
                for a in range(4):
                    if a not in valid_actions:
                        masked_q[a] = -1e9
                action = int(np.argmax(masked_q))

            # ---- Step environment ----
            next_envstate, reward, status = qmaze.act(action)

            # ---- Store transition ----
            episode = [envstate, action, reward, next_envstate, status]
            experience.remember(episode)

            # Move to next state
            envstate = next_envstate

            # ---- Train from replay ----
            try:
                inputs, targets = experience.get_data(data_size=data_size, gamma=gamma)
            except TypeError:
                # Fallback in case GameExperience.get_data doesn't accept gamma
                inputs, targets = experience.get_data(data_size=data_size)

            if inputs is not None and len(inputs) > 0:
                history = model.fit(inputs, targets, epochs=1, verbose=0)
                # Keep latest loss for logging
                loss = history.history.get('loss', [loss])[-1]

            # ---- Terminal? record result ----
            if status in ('win', 'lose'):
                win_history.append(1 if status == 'win' else 0)
                break

        # ---- Epsilon decay per episode ----
        epsilon = max(eps_min, epsilon * eps_decay)

        # ---- Stats & logging ----
        recent = win_history[-hsize:] if len(win_history) >= hsize else win_history
        win_rate = float(np.mean(recent)) if recent else 0.0
        n_episodes = len(win_history)

        dt = datetime.datetime.now() - start_time
        t = format_time(dt.total_seconds())
        template = ("Epoch: {:05d}/{:d} | Loss: {:.4f} | Episodes: {:d} | "
                    "Win count: {:d} | Win rate: {:.3f} | epsilon: {:.3f} | time: {}")
        print(template.format(epoch, n_epoch - 1, loss, n_episodes,
                              sum(win_history), win_rate, epsilon, t))


    # ---- Wrap up ----
    dt = datetime.datetime.now() - start_time
    seconds = dt.total_seconds()
    t = format_time(seconds)
    print("n_epoch: %d, max_mem: %d, data: %d, time: %s" % (epoch, max_memory, data_size, t))
    return seconds


# This is a small utility for printing readable time strings:
def format_time(seconds):
    if seconds < 400:
        s = float(seconds)
        return "%.1f seconds" % (s,)
    elif seconds < 4000:
        m = seconds / 60.0
        return "%.2f minutes" % (m,)
    else:
        h = seconds / 3600.0
        return "%.2f hours" % (h,)
```
